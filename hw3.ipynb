```python
#Wordü kod üzerinde değiştirerek yeni oyun oynanabilir.
word = "python course"
word = word.upper()
newWord = ""
count = 1
hangman = 0

def searchWord(letter):
    check = False
    for i in range(len(word)):
        if(word[i] == letter):
            global newWord
            newWord = newWord[:i] + letter + newWord[(i+1):]
            check = True
    return check

def searchNewWord(letter):
    check = False
    for i in range(len(newWord)):
        if(newWord[i] == letter):
            check = True
    return check 


for i in range(len(word)):
    if(word[i] ==" "):
        newWord+=" "
        count+=1
    else:
        newWord+="_"
print(">> ",newWord)
print(">> {} word".format(count))

while word != newWord:
    
    guessLetter = input("Please enter your guess: ").upper() 
    if searchNewWord(guessLetter):
        print("This input has already been used! Try Again..")
    else:
        if(searchWord(guessLetter)):
            print("Right Guess")
        else:
            print("Wrong Guess")
            hangman += 10
            print("Hang The Man %{}..".format(hangman))
            if hangman == 100:
                print("Failed..")
                break

    print("\n",newWord,"\n")

    
```
